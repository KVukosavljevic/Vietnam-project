```python
# Extracting all data files
f_n_1 = glob.glob(sh_path + "*\\*.csv")
f_n_2 = glob.glob(sh_path + "*\\*\\*.csv")
file_names_sh = []
file_names_sh.extend(f_n_1)
file_names_sh.extend(f_n_2)

print('All shimmer data file names:')
for name in file_names_sh : print(name)

f_n_1 = glob.glob(ppg_path + "*\\*.csv")
f_n_2 = glob.glob(ppg_path + "*\\*\\*.csv")
file_names_ppg = []
file_names_ppg.extend(f_n_1)
file_names_ppg.extend(f_n_2)

print('All PPG data file names:')
for name in file_names_ppg : print(name)
```

```python
IDs = []
IDs_L2 = []

hrv_mn = []
hrv_mn_L2 = []

hrv_sd = []
hrv_sd_L2 = []

hrv_ecgb = []
hrv_ecgb_L2 = []


for file in file_names_sh[:1]:
        #### Extract patient ID and wrap in dictionary df['ID'], extract necc chs df['ECG_LL']
        try:
            df = pd.read_csv(file, delimiter='\t', low_memory=False, header=1)

            # Extracting ECG_LL the one with mV
            data_ch1 = df['ECG_LL-RA_24BIT.1']
            data_ch1 = data_ch1.to_numpy()[:-3]
            data_ch1 = data_ch1[2:]
            data_ch1 = np.array([float(el) for el in data_ch1])

            print('Printing ECG_LL-RA_24BIT.1 data: ', data_ch1)

            # Raw ECG signal for one lead

            plt.rcParams['figure.figsize'] = [15, 5]  # Bigger images
            plt.rcParams['font.size'] = 14

            # Cleans signal with a 5th order high pass Butterworth filter with cut off at 0.5Hz and filters out mains frequency (set by default at 50Hz)
            #ecg_signals, info = nk.ecg_process(data_ch1, sampling_rate=100)

            #nk.ecg_plot(ecg_signals, sampling_rate=100)

            # Extracting HRV
            # Find peaks
            try:
                peaks, info = nk.ecg_peaks(data_ch1, sampling_rate=100)

                ID = int(file.replace(sh_path,"").split("\\", 1)[0].split("-", 2)[-1])

                hrv_time = nk.hrv_time(peaks, sampling_rate=100, show=False)
                hrv_freq = nk.hrv_frequency(peaks, sampling_rate=100, show=False, normalize=True)

                # get ID
                if "\\L2\\" in file :
                    # HRV
                    hrv_mn_L2.append(hrv_time['HRV_MeanNN'][0])
                    hrv_sd_L2.append(hrv_time['HRV_SDNN'][0])
                    
                    # ECG bands
                    a = hrv_freq['HRV_ULF'][0]
                    b = hrv_freq['HRV_VLF'][0]
                    c = hrv_freq['HRV_LF'][0]
                    d = hrv_freq['HRV_HF'][0]
                    e = hrv_freq['HRV_VHF'][0]
                    hrv_ecgb_L2.append([a, b, c, d, e])
                    
                    # ID
                    IDs_L2.append(ID)
                    
                    print('ID:', ID)
                    print('HRV ECG bands: ', hrv_ecgb[-1])
                else:    
                    # HRV
                    hrv_mn.append(hrv_time['HRV_MeanNN'][0])
                    hrv_sd.append(hrv_time['HRV_SDNN'][0])
                    
                    # ECG bands
                    a = hrv_freq['HRV_ULF'][0]
                    b = hrv_freq['HRV_VLF'][0]
                    c = hrv_freq['HRV_LF'][0]
                    d = hrv_freq['HRV_HF'][0]
                    e = hrv_freq['HRV_VHF'][0]
                    hrv_ecgb.append([a, b, c, d, e])
                    
                    #ID
                    IDs.append(ID)
                    
                    print('ID:')
                    print(ID)
                    print('HRV ECG bands')
                    print(hrv_ecgb[-1])
            except:
                print('Cant extract hrvtime for ', file)
                
        except:
            print('Cant load df for', file)
```
